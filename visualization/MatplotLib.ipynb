## **Bar Charts**

> **In short:** you use a bar chart when you want to compare discrete categories based on a specific value (like counts, averages, or sums).

1. **When to Use a Bar Chart**
- **Comparing Categories:** When you want to show how different groups (e.g., "Product Types," "Countries," or "Departments") stack up against each other.
- **Discrete Data:** Use them for data that has distinct labels. If your data is continuous (like temperature over time), a line chart is usually better.
- **Ranking:** They are excellent for showing "Top 10" lists because the human eye is very good at comparing the lengths of bars.
- **Frequency Distributions:** To show how often certain categorical values occur in a dataset.

2. **Choosing Your Orientation**

The direction of your bars actually matters for readability:
| Type | Best Used For... |
| :--- | :--- |
| **Vertical (Column)** | Standard comparisons with short category labels (e.g., years, months). |
| **Horizontal** | When you have **long category names** (e.g., "Customer Satisfaction Score") so the text doesn't overlap. |
| **Grouped (Clustered)** | Comparing multiple data series side-by-side within the same category. |
| **Stacked** | Showing the total value of a category while highlighting the sub-parts. |
| **100% Stacked** | Comparing the percentage/proportion of sub-groups rather than absolute numbers. |

---

3. **Variations for Complex Data**

- Sometimes a simple bar isn't enough. As a data science student, you’ll likely use these variations:
    - **Grouped (Clustered) Bar Chart:** Used to compare multiple series within the same category (e.g., "Sales" vs "Profit" per Region).
    - **Stacked Bar Chart:** Used when you want to show the total value of a category while also showing how it’s broken down into sub-parts.
    - **100% Stacked Bar Chart:** Used to show the relative percentage or proportion of sub-groups within each category, rather than absolute values.

![BarChart](../Images/BarChart.jpeg)

4. **Pro-Tips for Data Science**
- **Sort your bars:** Unless there is a natural order (like days of the week), always sort your bars from largest to smallest. It makes the "winner" instantly recognizable.
- **The "Zero" Rule:** Never start your Y-axis (or X-axis for horizontal charts) at anything other than 0. Truncating the axis exaggerates differences and is considered misleading data visualization.
- **Color with Purpose:** Don't use 10 different colors for 10 bars if they all represent the same metric. Use one color for the bars and perhaps a highlight color for the specific bar you're discussing.

````python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# plot the Bar Chartwith x-coordinates [0, 1, 2, 3, 4]
plt.bar(range(len(num_oscars)), num_oscars)

# Add The titles
plt.title("My Favorites Movies")

# Add y-label
plt.ylabel("Academy Awards")

# label x-axis with movies names
plt.xticks(range(len(movies)), movies)

plt.show()
````

### **Histograms**

> While bar charts are for categories, Histograms are for continuous, numerical data. You use a histogram when you want to see the "shape" of your data—its distribution, spread, and central tendency.

1. **When to Use a Histogram**
- **Visualizing Distribution:** To see if your data is "Normal" (Bell Curve), skewed to the left/right, or bimodal (two peaks).
- **Identifying Outliers:** To spot data points that fall far outside the main "clump" of data.
- **Analyzing Spread (Variance):** To see if your data is tightly packed around the mean or spread out widely.
- **Large Datasets:** When you have too many data points to plot individually (like a scatter plot) and need to group them into "bins."
![Histogram](../Images/Histogram.jpeg)

2. **Histogram vs. Bar Chart**

This is a common point of confusion for students. Here is how to tell them apart:
| Feature | Histogram | Bar Chart |
| :--- | :--- | :--- |
| **Data Type** | Quantitative (Continuous/Numerical) | Qualitative (Categorical/Discrete) |
| **Spacing** | Bars usually touch (shows continuity) | Space between bars (shows separation) |
| **X-Axis** | Ranges of values (Bins/Intervals) | Labels or Categories |
| **Order** | Must follow numerical order | Can be reordered (e.g., alphabetically or by value) |

---

3. **Key Concepts for Data Scientists**
- When you create a histogram in your notebook, you need to keep these two variables in mind:
    - **Bins:** This is how many intervals you divide your data into. Too few bins (over-smoothing) hides the detail; too many bins (under-smoothing) creates "noise."
    - **Density vs. Frequency:** By default, histograms show Frequency (count). However, you can set it to Density so the total area of the bars equals 1, which is useful for comparing distributions of different sample sizes.

4. **Anatomy of a Distribution**

As a data science student, you’ll use histograms to identify these patterns:
| Pattern | What it looks like | What it means |
| :--- | :--- | :--- |
| **Symmetric** | Bell-shaped / Gaussian | Mean, Median, and Mode are roughly equal. |
| **Right-Skewed** | Long tail on the right | Most data is small; a few very large values exist (e.g., household income). |
| **Left-Skewed** | Long tail on the left | Most data is large; a few very small values exist (e.g., age of death). |
| **Bimodal** | Two distinct peaks | Your data might contain two different groups (e.g., heights of men and women mixed together). |

````python
from collections import Counter
````

````python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Bucket grades by decile
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades) # Counter({80: 4, 90: 3, 70: 3, 0: 2, 60: 1})

plt.bar([x + 5 for x in histogram.keys()], # Shift bars right by 5
        histogram.values(),                # Give each bar its correct height
        10,                                # Give each bar a width of 10
        edgecolor=(0, 0, 0))               # Black edges for each bar

plt.axis([-5, 105, 0, 5])                   # x-axis from -5 to 105, y-axis from 0 to 5

plt.xticks([10 * i for i in range(11)])     # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("Grads of Students")
plt.title("Distribution of Exam 1 Grades")


plt.show()
````

````python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("The Times I heard someone say 'data science'")

# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False, style="plain")

# misleading y-axis only shows the part above 500
plt.axis((2016.5, 2018.5, 499, 506))

plt.title("Look at the 'Huge' Increase!")

plt.show()
````

## **Line Charts**

> A line chart is your go-to tool for visualizing trends over time (Time Series) or continuous relationships between two variables. While bar charts compare "snapshots" of different categories, line charts tell a story of "movement."

1. **When to Use a Line Chart**
- **Time Series Analysis:** When you want to see how a metric (like stock price, temperature, or website traffic) changes over days, months, or years.
- **Continuous Data:** When the data points have a logical connection to each other (e.g., measuring a plant's height every day).
- **Comparing Multiple Trends:** When you want to see if two different metrics are moving in the same direction (e.g., comparing "Ice Cream Sales" vs. "Temperature").
- **Identifying Volatility:** To see how much a variable fluctuates or "spikes" over a period.

2. **Line Chart vs. Bar Chart**

This is a critical distinction for your data science projects:
| Feature | Line Chart | Bar Chart |
| :--- | :--- | :--- |
| **Focus** | Movement, Trends, and Patterns | Magnitude and Individual Comparisons |
| **X-Axis** | Continuous (Time, Distance, Temperature) | Discrete (Categories, Names, Types) |
| **Connection** | Points are connected to show a "flow" | Bars are isolated to show "separation" |
| **Best For** | "How did we change from Jan to Dec?" | "Which month had the highest sales?" |

3. **Best Practices for Line Charts**

When coding these in Matplotlib or Seaborn, keep these rules in mind:
| Rule | Why it matters |
| :--- | :--- |
| **Don't Overcrowd** | Limit your chart to 3–4 lines. Too many lines create "spaghetti code" that is impossible to read. |
| **Consistent Intervals** | Ensure the time intervals on your X-axis are equal (e.g., don't jump from 1 day to 1 month randomly). |
| **Aspect Ratio** | Avoid making the chart too tall or too flat; it can make trends look more or less dramatic. |
| **Zero Baseline** | Unlike bar charts, line charts **do not** always need to start at zero; the focus is on the *change*. |

4. **Advanced Line Chart Types**

As you progress in data science, you'll encounter these variations:
| Variant | Purpose |
| :--- | :--- |
| **Area Chart** | A line chart with the area below filled in. Good for showing "Total Volume" over time. |
| **Step Chart** | Lines that move in "steps." Used for data that changes at specific intervals (e.g., interest rates). |
| **Multi-Axis** | Using two different Y-axes to compare two variables with different scales (e.g., Users vs. Revenue). |

---

> If your line chart looks "jagged" because of outliers, you can use a rolling mean (moving average) in Pandas to smooth the line:
```python
# Example: Smoothing a trend
df['smoothed_column'] = df['original_column'].rolling(window=7).mean()
plt.plot(df['date'], df['smoothed_column'])

````python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
totals_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# We can make multiple calls to plt.plot to show multiple series on the same chart
plt.plot(xs, variance, 'g-', label='variance')          # Green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2')       # reddot-dash line
plt.plot(xs, totals_error, 'b:', label='total error')   # blue dotted line

# loc=9 => mean "top center"
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The Bias-Variance Tradeoff")

plt.show()
````

## **Scatterplots**

> In data science, Scatterplots are the most important tool for identifying correlations and relationships between two numerical variables. While a line chart shows how one variable changes over time, a scatterplot shows how one variable affects another.

1. **When to Use a Scatterplot**
- **Correlation Analysis:** To see if two variables move together (e.g., "Hours Studied" vs. "Exam Score").
- **Outlier Detection:** To identify data points that don't follow the general trend of the dataset.
- **Cluster Identification:** To see if data points naturally group into distinct "blobs" (often used before applying K-Means clustering).
- **Checking Model Assumptions:** In Linear Regression, you use scatterplots of "Residuals" to check if your model is working correctly.
![Scatterplots](../Images/Scatterplot.jpeg)
---
2. **Interpreting the Relationship**

When looking at a scatterplot, you are primarily looking for the "direction" and "strength" of the relationship.
| Pattern | Description | What it Means |
| :--- | :--- | :--- |
| **Positive Slope** | Points move from bottom-left to top-right. | As X increases, Y also increases (e.g., Height and Weight). |
| **Negative Slope** | Points move from top-left to bottom-right. | As X increases, Y decreases (e.g., Car Age and Resale Price). |
| **No Pattern** | Points are scattered like a cloud. | There is no visible relationship between the variables. |
| **Non-Linear** | Points form a curve (U-shape or S-shape). | The relationship exists but is not a straight line. |

3. **Scatterplot vs. Other Charts**

As a student, you'll need to know why you'd pick this over a Line or Bar chart.
| Feature | Scatterplot | Line Chart | Bar Chart |
| :--- | :--- | :--- | :--- |
| **Data Types** | 2 Numerical Variables | 1 Time/Ord + 1 Numerical | 1 Categorical + 1 Numerical |
| **Goal** | Find Correlation | Show Trends | Compare Magnitudes |
| **Point Connection** | Never connected | Always connected | Not applicable |

4. **Advanced "Dimensions" in Scatterplots**

You can actually visualize more than just two variables (X and Y) on a single scatterplot by using aesthetic mapping:
| Dimension | Method | Usage |
| :--- | :--- | :--- |
| **3rd Variable** | **Color (Hue)** | Represent a category (e.g., Group A vs. Group B). |
| **4th Variable** | **Size** | Creates a Bubble Chart (e.g., Population size). |
| **Overplotting** | **Alpha** | Uses transparency to show density when points overlap. |

---

> If you have a massive dataset and your scatterplot looks like a solid black blob, use alpha=0.1 in your code. This makes the dots transparent, allowing you to see the "density" where most points are concentrated.
```python
plt.scatter(df['variable_x'], df['variable_y'], alpha=0.3, s=10)

````python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friends_count, minuts_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy=(friends_count, minuts_count),    # Put the label with its point
                xytext=(5, -5),                      # # Move 5 points horizontally, -5 points UP                  
                textcoords="offset points")          # Measurement unit is "points"
    
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("Nbr of friends")
plt.ylabel("daily minutes spent on the site")


plt.show()
````

> In data science, "comparable variables" are variables that use the same units or scale (e.g., height in cm vs. width in cm, or Math scores vs. Science scores).

> If you let Matplotlib choose the scale automatically, it will "zoom in" to fit the data perfectly into a rectangular box. This often stretches one axis more than the other, making a relationship look steeper or flatter than it actually is.

1. **The Core Problem: Visual Distortion**

When the X-axis and Y-axis have different scales (e.g., one inch on the screen represents 10 units on X, but 50 units on Y), the slope of your data is visually lied about.
| Scenario | What Matplotlib does (Default) | The Result |
| :--- | :--- | :--- |
| **Auto-Scale** | Stretches the axis with the smaller range to fill the plot area. | A 1:1 relationship (square) looks like a rectangle. |
| **Equal-Scale** | Forces 1 unit on X to be the same physical length as 1 unit on Y. | You see the *true* geometric relationship. |

---

2. **Why This Matters for Data Science**
- If you are analyzing variables that should be compared 1-to-1, an unequal scale can lead to wrong conclusions:
    - **Correlation Misinterpretation:** You might think a relationship is very "strong" (steep) when it's actually moderate.
    - **Clustering Errors:** In algorithms like K-Means, distance is everything. If your plot is stretched, the "distance" between points is visually deceptive.
    - **Isotropic Data:** If you expect a circular distribution of data, Matplotlib might make it look like an oval.

3. **The Solution: plt.axis('equal')**

To fix this, you tell Matplotlib to respect the geometry of the data.
| Command | Effect | Best Used For... |
| :--- | :--- | :--- |
| **`plt.axis('equal')`** | Ensures equal increments of data have equal screen length. | Standard comparisons (e.g., Height vs. Width). |
| **`plt.axis('square')`** | Makes the plot box a perfect square and scales axes equally. | Correlation plots where $X$ and $Y$ ranges are similar. |
| **`plt.tight_layout()`** | Adjusts padding between subplots and labels. | Preventing labels from clipping (Does NOT fix scale). |

---

- **The Golden Rule:** If your $X$ and $Y$ variables are measured in the same units, always force the axes to be equal. Otherwise, the "angle" of your correlation is a visual illusion created by the software, not the data.

```python
import matplotlib.pyplot as plt

# Plotting comparable variables
plt.scatter(heights, widths)

# The Fix:
plt.axis('equal') 

plt.show()

````python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)

plt.axis("equal")

plt.title("Axes Are Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")

plt.show()
````
