## **Vectors**

Think of vectors as the fundamental language of the field. Whether you're working with house prices, pixels in an image, or word embeddings in an LLM, you are essentially just manipulating vectors.

Here is a breakdown of what they are and the "rules of the game".

1. **What is a Vector?**
In Data Science, a vector is simply an ordered list of numbers.
- **Geometric view:** It's an arrow starting from the origin $(0,0)$ pointing to a specific coordinate. It has a magnitude (length) and a direction.
- **Data view:** It represents a "data point." For example, a house vector might be $[ \text{price}, \text{square footage}, \text{bedrooms} ]$.

2. **Fundamental Mathematical Rules**
To work with vectors, you need to follow specific algebraic rules. Let's assume we have two vectors:
- $$\mathbf{u} = \begin{bmatrix} u_1 \\ u_2 \end{bmatrix}, \mathbf{v} = \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}$$
- **A. Vector Addition**
    
You add vectors by adding their corresponding components. Geometrically, this is the "Tip-to-Tail" rule.
- $$\mathbf{u} + \mathbf{v} = \begin{bmatrix} u_1 + v_1 \\ u_2 + v_2 \end{bmatrix}$$

![Vectors](../Images/Vectors.jpeg)

- **B. Scalar Multiplication**

If you multiply a vector by a single number (a scalar, like $c=2$), you stretch or shrink the vector without changing its span (line).
- $$c\mathbf{v} = \begin{bmatrix} c \cdot v_1 \\ c \cdot v_2 \end{bmatrix}$$

- **C. The Dot Product (Crucial for Data Science!)**

The dot product results in a single number (scalar). It tells you how much two vectors "align".
-  $$\mathbf{u} \cdot \mathbf{v} = (u_1 \cdot v_1) + (u_2 \cdot v_2)$$
    - If the dot product is 0, the vectors are orthogonal (90 degrees apart).
    - In Machine Learning, we use this to calculate Cosine Similarity to see how similar two pieces of data are.
![vector projection](../Images/vector_projection.png)

3. **Measuring Vectors (Norms)**
- How "big" is your data point? We use the Norm (length). The most common is the L2 Norm (Euclidean distance):
    - $$\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}$$
**Why does this matter for you?**
In a typical Data Science project:
- **Features** are vectors.
- **Weights** in a model (like Linear Regression) are vectors.
- Training a model is basically just moving these vectors around in space until they point in the right direction to make accurate predictions.
> **Quick Tip:** When you start using Python, you'll use the NumPy library to handle these. A vector there is just a **numpy.array()**.

```python
import math
```

```python
def add_vectors(v: Vector, u: Vector) -> Vector:
    """add two equal vector by size, adding each element to the correspandent element in the second vector"""
    if len(v) != len(u): raise ValueError("The Vectors must be in the same leghts")
    
    return [v + u for v, u in zip(v, u)]


def sub_vectors(v: Vector, u: Vector) -> Vector:
    """sub two equal vector by size, subtracting each element to the correspandent element in the second vector"""
    assert len(v) == len(u), "The Vectors must be in the same leghts"
    
    return [v - u for v, u in zip(v, u)]


def sum_vectors(vectors: List[Vector]) -> Vector:
    """increasing dimension of vector by adding each position element by the others in rest vector"""
    # check of empty vectors
    assert vectors, "Please a vectors"

    # check of equal lenght
    vector_lenght = len(vectors[0])
    assert all(len(v) == vector_lenght for v in vectors ), "The Vectors must be in the same leghts"

    # the i-th element of the result is the sum of every vector[i]
    return [sum(v[i] for v in vectors)
                for i in range(vector_lenght)]


def scalar_multiply(v: Vector, c: float) -> Vector:
    """Multiply each element in vector by a scaler"""
    return [v_i * c for v_i in v]


def vector_mean(vectors: List[Vector]) -> Vector:
    """Computes the element-wise average"""
    n = len(vectors)
    return scalar_multiply(sum_vectors(vectors), 1/n)


def dot(v: Vector, w: Vector) -> float:
    """Computes v_1 * w_1 + ... + v_n * w_n"""
    assert len(v) == len(w), "The vector must be in same size"
    
    return sum(v_i * w_i for v_i, w_i in zip(v, w))

def sum_of_square(v: Vector) -> Vector:
    """Returns v_1 * v_1 + ... + v_n * v_n"""

    return dot(v, v)


def magnitude(v: Vector) -> float:
    """Returns the magnitude (or length) of v"""
    
    return math.sqrt(sum_of_square(v))


def squared_distance(v: Vector, w: Vector) -> float:
    """Computes (v_1 - w_1) ** 2 + ... + (v_n - w_n) ** 2"""

    return sum_of_square(sub_vectors(v, w))


def distance(v: Vector, w: Vector) -> float:
    """Computes the distance between v and w"""

    return magnitude(sub_vectors(v, w))
```

## **Matrices**

- In the context of data science and linear algebra, a **matrix** is a rectangular array of numbers, symbols, or expressions arranged in rows and columns.

- While a vector is a single list of numbers (a 1D array), a matrix is a 2D array. If a vector represents a point in space, a matrix often represents a **transformation** that moves or stretches that space.

---

- **Key Concepts for Data Science**

- **1. Representation of Data**
    - In your work with Databricks and data warehousing, you are effectively working with matrices constantly.
        * A **Dataset/DataFrame** is a matrix where each row is an observation (e.g., a specific customer) and each column is a feature (e.g., age, income, purchase history).

- **2. Matrix Multiplication (The Engine of AI)**
    - If you are moving into Agentic AI or working with Neural Networks, you will deal with matrix multiplication extensively. When a model "processes" data, it is essentially taking an input matrix (your data) and multiplying it by a weight matrix (the model's learned knowledge).

The result of this multiplication, $C = AB$, is a new matrix where every element is the dot product of a row from $A$ and a column from $B$.

- **3. Special Matrices**
    * **Identity Matrix ($I$):** The matrix equivalent of the number 1. Multiplying any matrix by the identity matrix leaves it unchanged.
    * **Transposed Matrix ($A^T$):** Swapping the rows and columns. This is used constantly in optimization algorithms to align dimensions for multiplication.
    * **Inverse Matrix ($A^{-1}$):** Used to "undo" a linear transformation (like solving for unknowns in a system of linear equations).

- **Matrix Notation**
    - A matrix $A$ with $m$ rows and $n$ columns is called an $m \times n$ matrix. You refer to an individual element using indices:
$$A_{ij}$$
- Where $i$ is the row number and $j$ is the column number.

```python
Matrix = List[List[Vector]]

A = [[1, 2, 3], # A has 2 rows and 3 columns
    [4, 5, 6]]

B = [[1, 2],    # B has 3 rows and 2 columns
    [3, 4],
    [5, 6]]
```

```python
from typing import Tuple, Callable
```

```python
def shape(A: Matrix) -> Tuple[int, int]:
    """Returns (number of rows of Matrix, number of columns of Matrix)"""
    num_rows = len(A)
    num_cols = len(A[0]) if A else 0

    return num_rows, num_cols


def get_row(A: Matrix, i: int) -> Vector:
    """Returns the i-th row of A (as a Vector)"""
    assert i < len(A), "the gien index is out of range"
    
    return A[i]


def get_column(A: Matrix, j: int) -> Vector:
    """Returns the j-th column of A (as a Vector)"""
    assert j < len(A[0]), "the gien index is out of range"
    
    return [ A_i[j] for A_i in A]


def make_matrix(num_row: int, num_cols: int, entry_func: Callable[[int, int], float]) -> Matrix:
    """
    Returns a num_rows x num_cols matrix
    whose (i,j)-th entry is entry_fn(i, j)
    """

    return [ [entry_func(i, j) for j in range(num_cols)] for i in range(num_row)]


def identity_matrix(n: int)-> Matrix:

    return make_matrix(n, n, lambda i, j: 1 if i == j else 0)
```
